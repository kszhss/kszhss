b. Korelasi antara Kolom duration dengan year_added

```python
# Membuat kolom baru untuk durasi dalam menit dan jumlah season
netflix_copy['duration_min'] = netflix_copy['duration'].apply(lambda x: int(x.split(' ')[0]) if "min" in x else None)
netflix_copy['duration_season'] = netflix_copy['duration'].apply(lambda x: int(x.split(' ')[0]) if "Season" in x else None)
netflix_copy.head(3)
```

```python
# Mengubah NaN menjadi 0 hanya pada kolom tertentu
netflix_copy['duration_min'] = netflix_copy['duration_min'].fillna(0)
netflix_copy['duration_season'] = netflix_copy['duration_season'].fillna(0)
```

```python
# Mengubah tipe data kolom duration_min dan duration_season menjadi Int32 (nullable integer)
netflix_copy['duration_min'] = netflix_copy['duration_min'].astype('int32')
netflix_copy['duration_season'] = netflix_copy['duration_season'].astype('int32')
```

2. Perbandingan Peminatan Berdasarkan Rating di Masing-masing Chanel tiap tayangan

```python
# Menghitung jumlah data berdasarkan nilai pada kolom 'rating'
rating_counts = netflix_copy['rating'].value_counts()

# Menampilkan hasil
print("Jumlah masing-masing kategori di kolom 'rating':")
rating_counts
```

```python
# Mengelompokkan data berdasarkan 'type' dan 'rating', kemudian menghitung jumlahnya
rating_counts_by_type = netflix_copy.groupby(['type', 'rating']).size().reset_index(name='count')

# Menampilkan hasil
print("Jumlah masing-masing kategori rating untuk setiap jenis acara:")
print(rating_counts_by_type)
```

```python
# Mengelompokkan data berdasarkan 'type' dan 'rating', kemudian menghitung jumlahnya
rating_counts_by_type = netflix_copy.groupby(['type', 'rating']).size().reset_index(name='count')

# Memisahkan data untuk TV Show dan Movie
tv_show_counts = rating_counts_by_type[rating_counts_by_type['type'] == 'TV Show'].set_index('rating')['count']
movie_counts = rating_counts_by_type[rating_counts_by_type['type'] == 'Movie'].set_index('rating')['count']

# Menggabungkan data TV Show dan Movie
combined_counts = pd.concat([tv_show_counts, movie_counts], axis=1, keys=['TV Show', 'Movie']).fillna(0)

# Membuat plot
plt.figure(figsize=(14, 7))
bars = combined_counts.plot(kind='bar', color=['pink', 'skyblue'], width=0.8, figsize=(14, 7))
plt.title('Jumlah Masing-Masing Kategori Rating untuk Setiap Jenis Acara')
plt.xlabel('Rating')
plt.ylabel('Count')

# Add annotations to each bar
for bar in bars.containers:
    bars.bar_label(bar, label_type='edge', fontsize=10, color='black')

plt.xticks(rotation=45)
plt.legend(title='Type')
plt.show()
```
